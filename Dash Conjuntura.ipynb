```python
import pandas as pd
import numpy as np
from datetime import date, time, datetime, timedelta
from dateutil.relativedelta import relativedelta
import dash
import dash_html_components as html
import dash_core_components as dcc
from dash_html_components import Div, H1, H2, P, H3
from dash_core_components import Graph, Slider, Checklist, Dropdown
from dash.dependencies import Input , Output
import dash_bootstrap_components as dbc
import plotly.express as px
import plotly.graph_objects as go
```

```python
#Bases
df_cambio = pd.read_excel('Câmbio.xlsx')
df_Pib = pd.read_excel ('PIB_%.xlsx')
df_export = pd.read_excel ('Exportação.xlsx')
df_ipca = pd.read_excel ('IPCA.xlsx')
df_selic = pd.read_excel ('Selic.xlsx')
df_gpd_world = pd.read_excel ('GPD_Trismetral.xlsx', sheet_name = 'GPD')
```

```python
df_gpd = pd.merge(df_Pib, df_gpd_world, on = ['Ano', 'Trimestre'])
df_gpd = df_gpd  [['Ano', 'Trimestre', 'PIB', 'gpd_world']]
df_gpd ['Ano Trimestre'] = df_gpd ['Ano'].astype(str) +" - " + df_gpd ['Trimestre']
df_gpd = df_gpd.query ('Ano>=2016')
df_gpd['GPD_World'] = df_gpd['gpd_world'].round(decimals=2)
df_gpd['PIB'] = df_gpd['PIB'].round(decimals=4)

df_gpd
```

```python
# merge dos dataframes

df_selic_ipca = pd.merge (df_selic,df_ipca, on = ['Período', 'Ano', 'Mês'])
df_selic_ipca = df_selic_ipca[['Período','Ano', 'Mês', 'Selic', 'IPCA']]
df_selic_ipca = df_selic_ipca.query ('Ano>=2019' )

df_cambio_export = pd.merge(df_cambio,df_export, on =['Ano', 'Trimestre'])
df_cambio_export = df_cambio_export[['Ano', 'Trimestre', 'cotacaoVenda', 'Export']]
df_cambio_export.rename(columns = {'cotacaoVenda':'Câmbio'}, inplace = True)

df_cambio_export = df_cambio_export.query ('Ano>=2016')
df_cambio_export['Câmbio'] = df_cambio_export['Câmbio'].round(decimals=2)
df_cambio_export ['Ano_trimestre'] = df_cambio_export['Ano'].astype (str) + ' - ' + df_cambio_export['Trimestre']

df_cambio_export
```

```python
df_cambio_export ['Ano_trimestre'] = df_cambio_export['Ano'].astype (str) + ' - ' + df_cambio_export['Trimestre']
```

```python
#tabelas plot
tb_selic_ipca = df_selic_ipca.query('Ano>=2022')

tb_cambio_export = df_cambio_export.query ('Ano>=2021')
tb_cambio_export['Export'] = tb_cambio_export['Export'].apply(lambda x : '{:,.2f}'.format(x))

tb_Pib = df_gpd.query ('Ano>=2021')

tb_Pib
```

```python
app = dash.Dash(__name__,  external_stylesheets = [dbc.themes.CYBORG])

fig_selic = go.Figure(layout = {"template": "plotly_dark"})
fig_selic.add_trace (go.Scatter (x = df_selic_ipca ['Período'] , y = df_selic_ipca ['Selic'], name = 'Selic'))

fig_selic.add_trace (go.Scatter (x = df_selic_ipca['Período'], y = df_selic_ipca['IPCA'], name = 'IPCA'))

fig_selic.update_xaxes (tickangle = 45, tickfont_size = 6, dtick = "M1", tickformat = "%b\n%Y",showgrid=False)
fig_selic.update_yaxes (tickfont_size = 10, showgrid=False)

fig_selic.update_layout (
    #title = 'Selic',
    autosize = True,
    margin = dict(l=40,r=40,t=40,b=40),
    legend = dict(
        orientation="h",
        yanchor="bottom",
        y=1.02,
        xanchor="right",
        x= 1.02)    
    )

fig_cambio_export = go.Figure (layout ={"template": "plotly_dark"})
fig_cambio_export.add_trace (go.Scatter (x = df_cambio_export['Ano_trimestre'], y = df_cambio_export['Câmbio'], 
                                         name = "Câmbio"))

fig_cambio_export.add_trace (go.Scatter (x = df_cambio_export['Ano_trimestre'], y = df_cambio_export['Export'], 
                                         name = "Export", yaxis="y2"))

fig_cambio_export.update_xaxes (tickangle = 45, tickfont_size = 6, showgrid=False)
fig_cambio_export.update_yaxes (tickfont_size = 10, showgrid=False)

fig_cambio_export.update_layout(
    yaxis = dict (title=dict(text="Câmbio"),side="left", showgrid=False),
    yaxis2 = dict (title=dict(text="Export"),side="right",overlaying="y", tickmode='array', showgrid=False),
    
    autosize= True,
    margin=dict(l=30,r=30,t=30,b=30),
    legend = dict(
        orientation="h",
        yanchor="bottom",
        y=1.02,
        xanchor="right",
        x= 1.02)
    )

fig_pib = go.Figure(layout = {"template": "plotly_dark"})
fig_pib.add_trace (go.Scatter (x = df_gpd['Ano Trimestre'], y = df_gpd['PIB'], name = 'Pib Brasil'))
fig_pib.add_trace (go.Scatter (x = df_gpd['Ano Trimestre'], y = df_gpd['gpd_world'], name = 'Pib mundo'))
fig_pib.update_xaxes (tickangle = 45, tickfont_size = 6, showgrid=False)
fig_pib.update_yaxes( tickfont_size = 10,showgrid=False)


fig_pib.update_layout(
       
    autosize = True,
    margin = dict (l=30,r=30,t=30,b=30),
    legend = dict(
        orientation="h",
        yanchor="bottom",
        y=1.02,
        xanchor="right",
        x= 1.02)
    )


tab_selic_ipca = go.Figure(layout ={"template": "plotly_dark"},data=[go.Table(
    header = dict (values= ['Ano', 'Mês','Selic', 'IPCA']),
    cells = dict (values = [tb_selic_ipca.Ano, tb_selic_ipca.Mês, tb_selic_ipca.Selic, tb_selic_ipca.IPCA]))
                           ])
tab_selic_ipca.update_layout(autosize = True,
                       title_text ="Selic / Ipca",
                      margin=go.Margin(l=20,r=20,t=50,b=40)
                      )

tab_cambio_export = go.Figure(layout = {"template": "plotly_dark"}, data = [go.Table(
    header = dict (values = ['Ano', 'Trimestre', 'Câmbio', 'Export']),
    cells = dict (values = [tb_cambio_export.Ano, tb_cambio_export.Trimestre, tb_cambio_export.Câmbio, 
                            tb_cambio_export.Export]))
                            ])

tab_cambio_export.update_layout (autosize= True,
                                 title_text = 'Câmbio / Exportação',
                                 margin = go.Margin (l=20,r=20,t=30,b=10)
                                 )

tab_pib = go.Figure(layout = {"template": "plotly_dark"}, data= [go.Table(
    header = dict (values = ['Ano', 'Trimestre', 'PIB', 'GPD World']),
    cells = dict (values = [tb_Pib.Ano, tb_Pib.Trimestre, tb_Pib.PIB, tb_Pib.GPD_World]))
                            ])
tab_pib.update_layout (autosize = True,
                      title_text = 'PIB',
                      margin = go.Margin(l=20,r=20,t=30,b=10)
                      )



#colocar em um eixo secundario
#fig_cambio_export.add_trace (go.Scatter (x = df_cambio_export['Ano_trimestre'], y = df_cambio_export['Export']))


fig_cambio_export.update_xaxes (tickangle = 45, tickfont_size = 6, dtick = "M1", tickformat = "%b\n%Y")        


app.layout = dbc.Container([
    dbc.Row([
        dbc.Col(
            html.H5("Conjuntura Econômica", style ={"margin-top": "10px", 'margin-botton': '40px'}), md =5        
        ),
        dbc.Col(md = 7)             
        
    ]),
    
    
   dbc.Row([
        dbc.Col(
                dcc.Graph (id = "tab_pib", figure = tab_pib, style = {"height":"45vh"}), md = 4
                            
            ),
            dbc.Col(
                dcc.Graph(id = "fig_pib", figure = fig_pib, style = {"height":"45vh"}),md =8
                
            )   
        
    ]),   
    
    dbc.Row([
        dbc.Col(
                dcc.Graph (id = "tab_cambio_export", figure = tab_cambio_export, style = {"height":"45vh"}), md = 4
                            
            ),
            dbc.Col(
                dcc.Graph(id = "cambio_export", figure = fig_cambio_export, style = {"height":"45vh"}),md =8
                
            )   
        
    ]),
    dbc.Row([
         dbc.Col(
                dcc.Graph (id = "tab_selic_ipca", figure = tab_selic_ipca, 
                           style = {"height":"45vh", "margin-botton": "25px"}),md = 4
                            
            ),
            dbc.Col(
                dcc.Graph(id = "selic_grafico", figure = fig_selic, style = {"height":"45vh", "margin-botton": "25px"}),md =8
                
            )       
        
    ])
    
           
    
], fluid = True)
                  
if __name__ == "__main__":
    app.run_server(debug = True, use_reloader=False)
       
```
